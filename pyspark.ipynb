```python
from pyspark.sql import SparkSession
# iniciando uma SessĂŁo do Spark
sc = SparkSession.builder.master('local[*]').getOrCreate()
# Verificando se a sessĂŁo foi criada
sc

dados_seguranca = sc.read.csv(
  path = "sample_data/ES_segDEZ2020.csv",
  inferSchema = True,
  header = True,
  sep = ';',
  encoding = "UTF-8")
type(dados_seguranca)


dados_seguranca.show()

dados_seguranca.printSchema()

dados_seguranca.select('DATA').printSchema()

from pyspark.sql.functions import col, count, when
dados_seguranca.select([count(when(col(c).isNull(), c)).alias(c) for c in
dados_seguranca.columns]).show()

from pyspark.sql.types import *
dados_seguranca1 = dados_seguranca.withColumn("IDADE", dados_seguranca["IDADE"].cast(IntegerType()))

for col in dados_seguranca1.dtypes:
  print(col[0]+"- "+col[1])



  from pyspark.sql.functions import *
rows = dados_seguranca.count()
print(f"Rows in Dataframe: {rows}")

from pyspark.sql.functions import *
dados_seguranca.select(count(dados_seguranca.IDADE)).show()

from pyspark.sql.functions import *
dados_seguranca.select(count(dados_seguranca.SEXO)).show()

from pyspark.sql.functions import col, avg
dados_seguranca_limpo = dados_seguranca.filter(col("IDADE").rlike("^[0-9]+$")) \
.withColumn("IDADE", col("IDADE").cast("int"))
dados_seguranca_limpo.select(avg("IDADE")).show()

from pyspark.sql.functions import stddev
dados_seguranca_limpo.select(stddev("IDADE")).show()

from pyspark.sql.functions import avg, stddev
dados_seguranca_limpo.select(
avg("IDADE").alias("media_idade"),
stddev("IDADE").alias("desvio_padrao")
).show()

dados_seguranca_limpo.select("IDADE").describe().show()

from pyspark.sql.functions import avg
dados_seguranca_limpo.groupBy("SEXO").avg("IDADE").show()

from pyspark.sql.functions import avg
dados_seguranca_limpo.groupBy("INCIDENTE").avg("IDADE").show()

dados_seguranca_limpo.groupBy("IDADE").count().orderBy("IDADE").show()
```
