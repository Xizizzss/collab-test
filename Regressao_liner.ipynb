```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([9714, 3728, 6062 , 8845, 8378 ])
```

```python
y = np.array([3011, 1305, 1879, 2654, 2849])
```

```python
coeficiente_angular, intercepto = np.polyfit(x, y, 1)
```

```python
print(f"Coeficiente angular: {coeficiente_angular}")
print(f"Intercepto: {intercepto}")
```

```python
y_pred = coeficiente_angular * x + intercepto
```

```python
print(y_pred)
```

```python
plt.scatter(x, y, label='Dados Originais')

plt.plot(x, y_pred, color='red', linewidth=2, label='linha de regressão')

plt.xlabel('Variável X')
plt.ylabel('Variável Y')
plt.legend()
plt.title('Regressão Linear Simples')
plt.grid(True)
plt.show()
```
